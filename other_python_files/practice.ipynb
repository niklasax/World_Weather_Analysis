```python
import pandas as pd
```

```python
donors_df = pd.read_csv("donors2008.csv")
donors_df
```

```python
avg_amt_state = donors_df.groupby(["State"]).mean()["Amount"]
avg_amt_state
```

```python
avg_amt_state_df = pd.DataFrame({"average amount by state":avg_amt_state})
avg_amt_state_df
```

```python

avg_amt_state_df.loc[avg_amt_state_df["average amount by state"] >= 750].count()
```
